```python

import numpy as np
import pandas as pd

# Set random seed for reproducibility
np.random.seed(42)

# Generate 1000 samples of user behavior data
num_samples = 1000
data = {
    "tab_switches": np.random.randint(0, 10, num_samples),  # 0-10 tab switches
    "idle_time": np.random.randint(0, 500, num_samples),  # 0-500 sec idle
    "mouse_movement": np.random.randint(50, 7000, num_samples),  # 50-7000 movements
    "clicks": np.random.randint(1, 150, num_samples),  # 1-150 clicks
    "typing_speed": np.random.randint(5, 150, num_samples),  # 5-150 WPM
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Introduce noise by adding random variations
noise_factor = 0.1
df["idle_time"] += np.random.normal(0, noise_factor * df["idle_time"].std(), num_samples)
df["mouse_movement"] += np.random.normal(0, noise_factor * df["mouse_movement"].std(), num_samples)
df["clicks"] += np.random.normal(0, noise_factor * df["clicks"].std(), num_samples)
df["typing_speed"] += np.random.normal(0, noise_factor * df["typing_speed"].std(), num_samples)

# Ensure values remain within realistic bounds
df = df.clip(lower=0)

# Introduce anomalies (e.g., extremely high tab switches or zero activity)
anomaly_indices = np.random.choice(df.index, size=50, replace=False)
df.loc[anomaly_indices, "tab_switches"] = np.random.randint(15, 30, size=len(anomaly_indices))

# Define risk levels with better balancing
def assign_risk(row):
    if row["tab_switches"] > 8 or row["mouse_movement"] < 200:  # More than 8 switches OR very low mouse movement → High Risk
        return "High"
    elif row["clicks"] < 20 or row["typing_speed"] < 40:  # Low clicks or slow typing → Medium Risk
        return "Medium"
    else:
        return "Low"

# Apply risk function
df["risk_label"] = df.apply(assign_risk, axis=1)

# Ensure classes are balanced
min_class_count = df["risk_label"].value_counts().min()
df_balanced = df.groupby("risk_label").apply(lambda x: x.sample(min_class_count, random_state=42)).reset_index(drop=True)

print(df_balanced["risk_label"].value_counts())  # Check class distribution
print("New randomized dataset generated successfully!")
```
